__Cleaning Cost__

```python
cost = []
for i in listings['cost'].values:
    try:
        val = int(re.findall('\w+',i)[0])
    except:
        print(i)
        val = i
    cost.append(val)
```

```python
cost = pd.Series(cost, name = 'cost')
```

```python
cost.head()
```

__Cleaning DateTime__

```python
date_time_ser = listings.datetime.values
date_time_list = []
for i in date_time_ser:
    date_time_dict = dict.fromkeys(['datetime_posted', 'datetime_update'])
    date_time_len = len(i)
    if date_time_len>2: print(i)
    if date_time_len==0:
        date_time_list.append(date_time_dict)
    else :
        datetime0 = datetime.strptime(i[0], '%Y-%m-%d %I:%M%p')
        if date_time_len==1:
            date_time_dict['datetime_posted'] = datetime0
        else :
            datetime1 = datetime.strptime(i[1], '%Y-%m-%d %I:%M%p')
            if datetime0<datetime1:
                date_time_dict['datetime_posted'] = datetime0
                date_time_dict['datetime_update'] = datetime1
            else:
                date_time_dict['datetime_posted'] = datetime1
                date_time_dict['datetime_update'] = datetime0
        date_time_list.append(date_time_dict)
```

```python
date_time_df = pd.DataFrame(date_time_list)
```

```python
date_time_df.head()
```

__Cleaning extra_info__

```python
ex_info = listings.extra_info.values.tolist()
```

```python
def extract_size(string):
    ba = re.findall('([\w\.]+)Ba', string)
    br = re.findall('([\w\.]+)BR', string)
    sz = re.findall('([\w]+)ft2', string)
    try:
        av = re.findall('available\s*([\w ]+)', string)[0]
    except:
        av = None
    size = []
    for i in [br, ba, sz]:
        try: 
            k = float(i[0])
        except:
            k = None
        size.append(k)
    size.append(av)
    return(size)
```

```python
def clean_extra_info(string):
    str_list = string.strip().split('\n\n')
    Br, Ba, size, available = extract_size(string)
    if len(str_list)>0:
        flag = True if any(x in string for x in ['Ba', 'BR', 'ft2', 'available']) else False
        if flag:
            other = ' '.join([i.replace(' ', '-') for i in str_list[1:]])
        else:
            other = ' '.join([i.replace(' ', '-') for i in str_list])
    else: other = ''
    return([Br, Ba, size, available, other])         
```

```python
extra_info_dict = []
for i in ex_info:
    Br, Ba, size, available, other = clean_extra_info(i)
    extra_info_dict.append({'bedroom': Br, 'bathroom': Ba, 'size_house': size, 'availability': available, 'other_info': other})
```

```python
extra_info_df = pd.DataFrame(extra_info_dict)
```

```python
extra_info_df.head()
```

__Tokenizing other_info__

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus_other_info = extra_info_df['other_info'].values
```

```python
vocab_other_info = []
for i in ' '.join(corpus_other_info.tolist()).split(' '):
    if i == '' or len(re.findall('[0-9]+', i))>0: continue
    if i not in vocab_other_info: vocab_other_info.append(i)
    #unique_extra_info_dict[i] = unique_extra_info_dict.get(i,0)+1
```

```python
other_info_token = CountVectorizer(vocabulary=vocab_other_info, tokenizer=lambda x: x.split(' ')).fit(corpus_other_info)
```

```python
other_info_token = pd.DataFrame(other_info_token.transform(corpus_other_info).toarray(), columns=other_info_token.get_feature_names())
```

__Clean last_pos_time__

```python
last_time = listings['last_pos_time'].values
```

```python
last_time_list = []
for i in last_time:
    try:
        t = datetime.strptime(i, '%Y-%m-%d %H:%M')
    except:
        print(i)
        t = None
    last_time_list.append(t)
```

```python
last_time_df = pd.Series(last_time_list, name = 'last_time_pos')
```

```python
last_time_df.head()
```

__Cleaning Latitude and Longitude__

```python
latlon = listings[['latitude', 'longitude']].astype('float64')
```

__Cleaning place__

```python
place_ser = listings['place'].values
place_list = []
for place in place_ser:
    try:
        p = re.findall('[\w ]+', place)[0]
    except:
        p = None
    place_list.append(p)
```

```python
place_df = pd.Series(place_list, name='place')
place_df.head()
```

__Concatenate all__

```python
listings_clean = pd.concat([latlon, place_df, cost, date_time_df, last_time_df, extra_info_df, other_info_token], axis = 1)
```
